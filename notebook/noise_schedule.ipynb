```python
import numpy as np
from improved_diffusion.gaussian_diffusion import get_named_beta_schedule
from matplotlib import pyplot as plt
```

```python
def calculate_tilde_alpha(betas):
    alpha = 1. - betas
    return np.cumprod(alpha)
```

```python
def brand_new_linear_beta(b_min, b_max, num_diffusion_timesteps):
    betas = [b_min / num_diffusion_timesteps]
    for i in range(num_diffusion_timesteps - 1):
        betas.append(b_min / num_diffusion_timesteps + (i+1) / (num_diffusion_timesteps*(num_diffusion_timesteps-1)) * (b_max - b_min))
    return np.array(betas)
```
