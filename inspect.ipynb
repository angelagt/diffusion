```python
import matplotlib.pyplot as plt 
import numpy as np

images_npz = np.load('./orig_samples/samples_16x32x32x3.npz')
images_fn = images_npz.files[0]

images_arr = images_npz[images_fn]

fig, ax = plt.subplots(4,4, figsize=(20, 20))
for i in range(16):
    r = int(i / 4)
    c = int(i % 4)
    ax[r, c].imshow(images_arr[i])
```
